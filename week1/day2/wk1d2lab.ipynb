2. Select the columns which contain the information from years 2002 to 2008. Ensure you keep the Index and State variables as well.

```python
states_subset = states.loc[:, "Index":"Y2008"]
```

```python
states_subset
```

3. Rename the Index column to first_letter

```python
states_subset = states_subset.rename(columns={"Index":"first_letter"})
```

```python
states_subset
```

4. Find the seven states which start with an A,B, or C.

```python
conditionA = states_subset.first_letter == 'A'
```

```python
conditionB = states_subset.first_letter == 'B'
```

```python
conditionC = states_subset.first_letter == 'C'
```

```python
combined_condition = conditionA | conditionB | conditionC
```

```python
seven_states = states_subset[combined_condition]
```

5.Find which of these seven states have the greatest income in 2008.

```python
seven_states.loc[:, ["State","Y2008"]].sort_values("Y2008", ascending=False).head(1)
```

6. Calculate the change in income for each state of these seven states (states starting with A,B, or C) between 2002 and 2008. Save the income change in a new column called income_change.

```python
seven_states = seven_states.assign(income_change = seven_states.Y2008 - seven_states.Y2002)
```

7. Find which of these seven states income change is greater than 500,000 (i.e. an increase of 500,000 or decrease of 500,000).

```python
pos = 500000
neg = -500000
```

```python
bool_check = (seven_states["income_change"] > pos) | (seven_states["income_change"] < neg)
```

```python
seven_states[bool_check]
```

8. Calculate the mean and median income change for all seven states.

```python
round(seven_states["income_change"].mean())
```

```python
round(seven_states["income_change"].median())
```
