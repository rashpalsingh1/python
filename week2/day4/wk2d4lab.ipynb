10) Using uk_diet, create a new column that contains the difference in total calories between a year and the year before. Then find the year where there was the biggest positive difference and the biggest negative difference.

```python
uk_diet["shifted"] = uk_diet["avg_daily_kcals"].shift(periods=1)
```

```python
uk_diet["difference"] = uk_diet["avg_daily_kcals"] - uk_diet["shifted"]
```

```python
uk_diet.sort_values(by="difference", ascending=False).head(1) #largest positive difference
```

```python
uk_diet.sort_values(by="difference", ascending=True).head(1) #largest negative difference
```

11) Back to diet_comp_clean again. For every year between 1990 and 2000, find which country got the most average daily calories from alcoholic beverages, i.e. you want to end up with one country per year.

```python
diet_comp_clean_year_sort = diet_comp_clean.sort_values(by="year") # sort by year
```

```python
diet_comp_clean_year_sort = diet_comp_clean_year_sort[(diet_comp_clean_year_sort["year"] > 1990) & (diet_comp_clean_year_sort["year"] < 2001)]
# filter the required date range
```

```python
diet_comp_clean_year_sort = diet_comp_clean_year_sort[diet_comp_clean_year_sort["kcal_source"].str.contains("bever", case=False)]
#filter beverages only
```

```python
(diet_comp_clean_year_sort.groupby(["year"])
.avg_daily_kcals.max().reset_index().merge(diet_comp_clean_year_sort, on=["avg_daily_kcals", "year"]))
#groupby year and find max kcals, then merge this with DF in previous step
```
