5. Read in hat_observations and separate observation into two columns, hat_colour and hat_type.

```python
hat = pd.read_csv("hat_observations.csv")
```

```python
hat_split = hat["observation"].str.split(pat=",", expand=True)
```

```python
hat["hat_colour"] = hat_split.loc[:, 0]
```

```python
hat["hat_type"] = hat_split.loc[:, 1]
```

```python
hat = hat.drop(columns="observation") # remove redundant column
```

6. Unite day, month, and year columns into a column called date using a suitable separator. Then find the date where the most berets were observed.

```python
hat["date"] = hat["day"].astype("str") + "_" + hat["month"].astype("str") + "_" + hat["year"].astype("str")
```

```python
(hat.groupby(["date", "hat_type"]).agg({"observation_count":"sum"})
.sort_values(by="observation_count",ascending=False).filter(like= "beret", axis = 0).head(1))
```
