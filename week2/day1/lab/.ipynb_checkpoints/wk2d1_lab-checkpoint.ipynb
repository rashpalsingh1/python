```python
import pandas as pd
```

1. List all the roles, along with the actors who played each role in The Little Mermaid. (In the data as “Little Mermaid, The”)

```python
movies = pd.read_csv("movies.csv")
```

```python
actors = pd.read_csv("actors.csv")
```

```python
roles = pd.read_csv("roles.csv")
```

```python
movies_roles = movies.merge(roles, left_on="id", right_on="movie_id")
```

```python
movies_roles[movies_roles["name"] == "Little Mermaid, The"]
```

2. Find the actor who has appeared in most films. List the films they appeared in, the year the film was released and their role in the film.

```python
most_popular_actor = actors.sort_values(by="film_count", ascending=False).head(1)["id"].values[0]
```

```python
movies_roles.loc[movies_roles["actor_id"] == most_popular_actor]
```

3. For all actors who have appeared in at least two films, find the average rating of the films they appeared. Find the five actors with the highest average rank of films they appeared in.

```python
all_combi = movies_roles.merge(actors, left_on="actor_id", right_on="id")
```

```python
all_combi_2 = all_combi[all_combi.film_count > 1].sort_values(by="film_count", ascending=False)
```

```python
(all_combi_2.groupby(["actor_id", "last_name", "first_name"])
 .agg({"rank":"mean"}).sort_values(by="rank", ascending=False)).head(5)
```
