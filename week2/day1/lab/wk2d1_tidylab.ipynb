1. Load in the country_data.csv file and take a look at it. What needs changed in this to make it tidy? Which columns need made into long format, and which need spread to rows? Do the appropriate pivoting of your columns.

```python
country = pd.read_csv("country_data.csv").clean_names()
```

```python
country = country.drop(columns="unnamed_0")
```

```python
country_updated = country.melt(id_vars=["country", "type", "code_letter", "code_numb"], var_name="year", value_name="numbers")
```

```python
country_long = country_updated.pivot_table(index = ["country","code_letter", "code_numb", "year"], columns="type", values="numbers").reset_index()
```

```python
country_long
```

2. Separate the country variable so you have the name in one column and the number in another.

```python
country_split = country_long["country"].str.split(pat=".", expand=True)
```

```python
country_split = country_split.drop(columns= 0)
```

```python
country_long["country"] = country_split
```

3. Unite the code_letter and code_numb variables so that they are both in one column, with no spaces separating them.

```python
country_long
```

```python
country_long["letter_numb"] = country_long["code_letter"] + country_long["code_numb"].astype("str")
```

```python
country_long = country_long.drop(columns= ["code_letter", "code_numb"])
```

```python
year_split = country_long["year"].str.split(pat="y", expand=True)
```

```python
year_split = year_split.drop(columns= 0)
```

```python
country_long["year"] = year_split
```
